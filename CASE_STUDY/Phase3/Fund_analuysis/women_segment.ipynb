```python
import pandas as pd
import numpy as np
```

```python
xls = pd.ExcelFile('completed_customer_complete_info.xlsx')
customer = pd.read_excel(xls, 'Sheet1')
```

```python
customer.describe()
```

```python
Customer=customer
```

```python
Customer.columns
```

```python
customer.shape
```

```python
import matplotlib.pyplot as plt

Customer_filtered = Customer[(Customer["Gender"] == "Female") & 
                            (Customer["Number_Of_Dependents"].isin([0,1,2])) & 
                            (Customer["oid_min_asset"] != 4) & 
                            (Customer["Risk_Score"].between(2,4))]

plt.boxplot(Customer_filtered["Retirement_asset"])
plt.title("Boxplot of Retirement_asset before outlier removal")
plt.xlabel("Retirement_asset")
plt.ylabel("Value")
plt.show()

Q1 = np.percentile(Customer_filtered["Retirement_asset"], 25)
Q3 = np.percentile(Customer_filtered["Retirement_asset"], 75)

IQR = Q3 - Q1

print("Q1 is", Q1)
print("Q3 is", Q3)
print("IQR is", IQR)

low_lim = Q1 - 1.5 * IQR
up_lim = Q3 + 1.5 * IQR

print("The lower limit for outliers is", low_lim)
print("The upper limit for outliers is", up_lim)

outliers = Customer_filtered[(Customer_filtered["Retirement_asset"] < low_lim) | (Customer_filtered["Retirement_asset"] > up_lim)]

print("The outliers of Retirement_asset column are:")
print(outliers)

Customer_clean = Customer_filtered[(Customer_filtered["Retirement_asset"] >= low_lim) & (Customer_filtered["Retirement_asset"] <= up_lim)]

plt.boxplot(Customer_clean["Retirement_asset"])
plt.title("Boxplot of Retirement_asset after outlier removal")
plt.xlabel("Retirement_asset")
plt.ylabel("Value")
plt.show()

print("The shape of the cleaned Customer dataframe is:", Customer_clean.shape)

print("The summary statistics of the cleaned Customer dataframe are:")
print(Customer_clean.describe())
```
