```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats
```

```python
xls = pd.ExcelFile('data_schema_v3.xlsx')
customer = pd.read_excel(xls, 'Customers')
customer_assets = pd.read_excel(xls, 'Customer_Assets')
asset_obj=pd.read_excel(xls,'Asset_Objectives')
customer_ans=pd.read_excel(xls,'Customer_Answers')
questions=pd.read_excel(xls,'Questions')
answers=pd.read_excel(xls,'Answers')
risk_prof=pd.read_excel(xls,'Risk_Profile')
asset_classes=pd.read_excel(xls,'Asset_Classes')
fee=pd.read_excel(xls,'Fee_Structures')
```

```python
### outlier analysis based on total_asset vs no of dependents for ASSET ID 3

# Read the customers and customer_assets tables from the database
customers = pd.read_excel(xls, 'Customers')
customer_assets = pd.read_excel(xls, 'Customer_Assets')
# Create views DA1 and DA2 by selecting and filtering columns
DA1 = customers[["customer_ID", "Number_Of_Dependents"]]
DA2 = customer_assets[customer_assets["Asset_Objective_ID"] == 3]

DA1.columns = DA1.columns.str.lower()
DA2.columns = DA2.columns.str.lower()
# Create view DA3 by joining DA1 and DA2 on customer_ID and filtering by number_of_dependents = 1
DA3 = DA1.merge(DA2, on="customer_id", how="inner")
#DA3 = DA3[DA3["number_of_dependents"] == 1]
```

```python
### plot scatter plot

import matplotlib.pyplot as plt

# Create a scatter plot of total vs number_of_dependents
plt.scatter(DA3["total"], DA3["number_of_dependents"])

# Add labels and title
plt.xlabel("Total")
plt.ylabel("Number of Dependents")
plt.title("Scatter Plot of Total vs Number of Dependents")

# Show the plot
plt.show()
```

```python
z = stats.zscore(DA3["total"])

cleaned34 = DA3 [np.abs (z) < 3]
outliers3=DA3 [np.abs (z) >= 3]
#fig, (ax1, ax2) = plt.subplots(1, 2)
print(len(outliers3))
```
